```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from cleanupFile import cleanUp, custom_acc_type
plt.rcParams['figure.figsize'] = (30,15)
plt.rcParams.update({'font.size': 15})
```

```python
df = pd.read_csv("./Dataset5.csv")
cleanUp(df)
df
```

```python
df['MAJOR ACCIDENT TYPE 1'] = df.apply(custom_acc_type, column=1 , axis=1)
df.shape
```

```python
df1 = df[['CONTRIBUTING FACTOR VEHICLE 1','MAJOR ACCIDENT TYPE 1']]
df1.rename(columns={"MAJOR ACCIDENT TYPE 1":'parent',"CONTRIBUTING FACTOR VEHICLE 1":'label'},inplace=True)
df1.columns
```

```python
df4 = df1.groupby('parent').size()
label=[]
values=[]
parent=[]
```

```python
for i in df4.index:
    print(i)
    df5 = df1[df1['parent']==i]
    df6 = df5.groupby('label').size()
    for j in df6.index:
        if j=="Driver Inexperience":
            continue
        label.append(j)
        values.append(df6[j])
        parent.append(i)
```

```python
print(label)
print(values)
print(parent)
```

```python
dic = {"label":label,"parent":parent,"values":values}

df=pd.DataFrame(dic)

df.to_csv('file.csv',index=False)
```
