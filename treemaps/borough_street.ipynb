```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from cleanupFile import cleanUp, custom_acc_type
plt.rcParams['figure.figsize'] = (30,15)
plt.rcParams.update({'font.size': 15})
```

```python
df = pd.read_csv("./Dataset5.csv")
cleanUp(df)
df
```

```python
df1 = df[['BOROUGH','ON STREET NAME']]
df1.dropna(subset=['BOROUGH'], inplace=True)
df1.dropna(subset=['ON STREET NAME'], inplace=True)

for i in range(len(df1.index)):
    df1.iloc[i,0] = df1.iloc[i,0].strip()
    df1.iloc[i,1] = df1.iloc[i,1].strip()


df4 = df1.groupby('BOROUGH').size()

label=[]
values=[]
parent=[]

for i in df4.index:
    print(i)
    df5 = df1[df1['BOROUGH']==i]
    df6 = df5.groupby('ON STREET NAME').size()
    for j in df6.index:
        label.append(j.strip())
        values.append(df6[j])
        parent.append(i.strip())
```

```python
print(label)
print(values)
print(parent)
```

```python
dic = {"label":label,"parent":parent,"values":values}

df=pd.DataFrame(dic)
df=df[df['values']>63]

df.to_csv('file1.csv',index=False)
```
