```python
import pandas as pd
import numpy as np
import re
```

```python
node_set = set([])
edge_list = []
```

```python
f = open('./node_link_dataset/download.tsv.convote/convote/out.convote')
lines = f.readlines()

nodes = {}

for line in lines[1:]:
    splitted_line = re.split(r'[ \n]', line)
    splitted_line.pop()
    edge_list.append({'Source':splitted_line[0], 'Target':splitted_line[1], 'Type':'Directed', 'Weight':int(splitted_line[2])})
    node_set.add(splitted_line[1])
    node_set.add(splitted_line[0])
```

```python
node_list = []
for i in node_set:
    node_list.append({'Id' : i, 'Label' : i})

nodes_df = pd.DataFrame(node_list)
nodes_df
```

```python
edges_df = pd.DataFrame(edge_list)
edges_df
```

```python
edges_df.to_csv("./node_link_dataset/edges.csv", index=False)
nodes_df.to_csv("./node_link_dataset/nodes.csv", index=False)
```
